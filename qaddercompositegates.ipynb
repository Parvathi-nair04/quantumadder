```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
sub_c=QuantumCircuit(4)
```

```python
sub_c.ccx(0,1,2)
sub_c.cx(0,1)
sub_c.ccx(1,3,2)
sub_c.cx(1,3)
```

```python
sub_inst=sub_c.to_instruction()
```

```python
#Part 1
```

```python
qc=QuantumCircuit(13,5)
```

```python
#Initializing A and B to |0010> and |1011>
qc.x(1)
qc.x(4)
qc.x(5)
qc.x(11)
```

```python
qc.append(sub_inst,[0,1,2,3])
```

```python
qc.append(sub_inst,[4,5,6,2])
qc.append(sub_inst,[7,8,9,6])
qc.append(sub_inst,[10,11,12,9])
```

```python
qc.measure(3,0)
qc.measure(2,1)
qc.measure(6,2)
qc.measure(9,3)
qc.measure(12,4)
```

```python
qc.draw('mpl')
```

```python
backend_sim = Aer.get_backend('qasm_simulator')
job_sim = backend_sim.run(transpile(qc, backend_sim), shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts(qc)
print(counts)
```

```python
plot_histogram(counts)
```
