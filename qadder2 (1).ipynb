```python
from qiskit import *
from qiskit.visualization import plot_histogram
qc=QuantumCircuit(13,5)
```

```python
#Initializing A and B to |0001> and |0011>
qc.x(0)
qc.x(1)
qc.x(5)
```

```python
#Quantum full adder 1
qc.ccx(0,1,2)
qc.cx(0,1)
qc.ccx(1,3,2)
qc.cx(1,3)
```

```python
#Quantum full adder 2
qc.ccx(4,5,6)
qc.cx(4,5)
qc.ccx(5,2,6)
qc.cx(5,2)
```

```python
#Quantum full adder 3
qc.ccx(7,8,9)
qc.cx(7,8)
qc.ccx(8,6,9)
qc.cx(8,6)
```

```python
#Quantum full adder 4
qc.ccx(10,11,12)
qc.cx(10,11)
qc.ccx(11,9,12)
qc.cx(11,9)
```

```python
qc.draw('mpl')
```

```python
qc.measure(3,0)
qc.measure(2,1)
qc.measure(6,2)
qc.measure(9,3)
qc.measure(12,4)
```

```python
#Final circuit
qc.draw('mpl')
```

```python
backend_sim = Aer.get_backend('qasm_simulator')
job_sim = backend_sim.run(transpile(qc, backend_sim), shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts(qc)
print(counts)
```

```python
plot_histogram(counts)
```
